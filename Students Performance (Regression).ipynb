# Linear Least Squares (LLS)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.model_selection import train_test_split
```

```python
x = np.random.uniform(1,15,150)
y = x + np.random.uniform(0.25,5,150)
```

```python
x
```

```python
y
```

## Generate continuous random dataset for the problem of X = study hours and Y = grade of students (if you study more, you will get a higher grade).

```python
plt.scatter(x, y, color='lightgreen')
plt.xlabel('Study')
plt.ylabel('Grade')
plt.show()
```

```python

x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.2 )
```

```python
x_train
```

```python
plt.scatter(x, y,color='lightgreen')
plt.plot(x_train , 0.4 * x_train ,color="red")
plt.xlabel('Study')
plt.ylabel('Grade')
plt.show()
```

```python
X_train =np.array( x_train)

X_train = X_train.reshape(-1,1)
```

```python
Y_train = np.array(y_train)

Y_train = Y_train.reshape(-1,1)
```

```python
# W = (X.T * X)^ -1 * X.T * Y
W = np.matmul(np.matmul(np.linalg.inv( np.matmul(X_train.T , X_train )) , X_train.T) , Y_train )
```

```python
W
```

```python
plt.scatter(x, y, color='lightgreen' , label= 'data')
#plt.plot(X_train , 0.4 * X_train , color="red")
plt.plot(X_train , W * X_train ,c ='blue' , label ='Random Slope')
plt.xlabel('Study')
plt.ylabel('Grade')
plt.show()
```

```python
x_train , y_train
```

```python
slope , intersept , r , p ,std_err = stats.linregress(x_train , y_train)

def scipyLR (X) :
    return slope * X + intersept

mymodel = list(map(scipyLR , x_train))

plt.scatter(x_train ,y_train , c='lightgreen')
plt.plot(x_train , mymodel , c='green')
plt.xlabel('Study')
plt.ylabel('Grade')
plt.show()
```

```python
plt.scatter(x, y, color='lightgreen' , label='data')
plt.plot(X_train , 0.4 * X_train , color="red" , label ='Random Slope')
plt.plot(X_train , W * X_train ,c ='blue' ,label ='LLS' )
plt.plot(x_train , mymodel , c='green' , label='scipy')
plt.xlabel('Study')
plt.ylabel('Grade')
plt.legend()
plt.show()
```
